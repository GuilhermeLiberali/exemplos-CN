```python
#Exemplo de gráfico em python para auxiliar na escolha do intervalo

#http://matplotlib.org/
import matplotlib.pyplot as plt
#https://numpy.org/
import numpy as np

def flinear(x):
    return x-0.5

def fquadratica(x):
    return x**2+1

def fcubica(x):
    return x**3-1

# Exemplo "Contínua"
a = -1.5
b = 1.5
plt.figure(1)
x = np.linspace(a,b,50)
plt.plot(x, flinear(x), 'r-.', label = 'Linear')
plt.plot(x, fquadratica(x), 'b-', label = 'Quadrática')
plt.plot(x, fcubica(x), 'g.', label = 'Cúbica')
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.title("Exemplo de gráfico")
plt.legend()
plt.show()
```
